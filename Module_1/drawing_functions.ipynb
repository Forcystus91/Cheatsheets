When develoing a model in machine learning, it is usually convenient to work with a toy dataset or to generate a synthetic dataset to test your model on.

```python
import numpy as np
import matplotlib.pyplot as plt
```

When you define a line in Math, you usually use a forumla like the following:

$$y = 2x +5$$

Then, to draw it, you usuaully assign two values to $x$ and see the corresponding value of $y$:

$$x\,y\\
0\,5\\
1\,15
$$

Now you can plot the two points on your cartesian plane and draw a line between them. We can do it in Python using plt.plot():

```python
plt.plot([0,1],[5,15])
```

But what if we want to draw a more complicated curve. Let's take $y=\sin{\theta}$

Looking at some values, we have:

$$x\,\,\,y\\
0\,\,\,0\\
\frac{\pi}{2}\,\,\,1\\
\pi\,\,\,0\\
\frac{3}{2\pi}\,-1\\
2\pi\,\,\,0\\
$$

```python
x = [0, np.pi/2, np.pi, 3/2*np.pi, 2*np.pi]
y = [0,1,0,-1,0]
```

```python
plt.plot(x,y)
```

It doesn't quite look like sin, so let's add some more points.

```python
x = [0, np.pi/4, np.pi/2, 3/4*np.pi, np.pi, 5/4*np.pi, 3/2*np.pi, 7/4*np.pi, 2*np.pi]
y = [0, np.sqrt(2)/2, 1, np.sqrt(2)/2, 0, -np.sqrt(2)/2, -1, - np.sqrt(2)/2, 0]
```

```python
plt.plot(x,y)
```

It's getting closer. If we could find 1000 points it would look better.

```python
x = np.linspace(0,2*np.pi, 1000)
x
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

We could extend the interval of x as well:

```python
x = np.linspace(-10*np.pi, 10*np.pi, 1000)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = np.linspace(-50,50, 1000)
y = x**2
plt.plot(x,y)
```
